```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
```

```python
iris=load_iris()
x=iris.data
y=iris.target
```

```python
import tensorflow as tf
from tensorflow import keras
m1=keras.models.Sequential()
m1.add(keras.layers.Dense(units=2,activation='relu',input_shape=(4, )))
m1.add(keras.layers.Dense(units=3,activation='relu'))
m1.add(keras.layers.Dense(units=3,activation='sigmoid'))
m1.summary()
```

```python
m1.compile(loss='sparse_categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
m1.fit(x,y)
test_loss,test_accuracy=m1.evaluate(x,y)
```
